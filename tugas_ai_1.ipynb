Pada tugas ini, saya menggunakan ***library* yang saya buat sendiri** untuk mengambil dari data https://api.kaltimprov.go.id secara *realtime*.

4. Untuk alur perintahnya, saya deskripsikan ditiap barisnya.


Catatan :
Untuk memudahkan proses demonstrasi, saya melampirkan file `api_auth.json` melalui waskita atau bisa diunduh melalui tautan :  https://drive.google.com/file/d/1J5Bjoh9zXuNcTwIzWnT9eeZ1G42TdGwa/view?usp=sharing



```python
# Install library apisamarinda dan pandas bila belum build-in
# %pip install apisamarinda pandas

# Import pustaka yang dibutuhkan, dalam hal ini saya menggunakan fungsi api dari apisamarinda dan pandas secara utuh
from apisamarinda import api
import pandas as pd

# Proses autentikasi menggunakan file credentials dari API Kaltim
request = api.ApiClient("/content/api_auth.json")

# Menentukan endpoint data atau informasi yang ingin digunakan
result = request.result('https://api.kaltimprov.go.id/api/generate/samarinda/diskominfo/berita')

# Mengkonversi hasil kedalam bentuk JSON (menggunakan pustaka pandas)
json = pd.read_json(result.text)

# Menampilkan hasil dalam bentuk tabular
display(json)
```

#### Melakukan Telaah Data

```python
# Menampilkan tipe data tiap kolom
json.dtypes
```

```python
# Contoh menghilangkan kolom yang tidak digunakan
json_tanpa_uuid = json.iloc[:,2:]
json_tanpa_uuid
```

```python
# Statistik Dasar tiap kolom
json_tanpa_uuid.describe()
```

```python
# Menampilkan semua statistik termasuk yang bukan numerikal
# Catatan : datetime_is_numeric dibutuhkan untuk menghilangkan warning karena fungsi .describe() akan segera deprecated, 
# dalam arti kedepannya ada kemungkinan fungsi .describe tidak akan mendukung butir data bertipe datetime
json_tanpa_uuid.describe(include='all', datetime_is_numeric=True)

# Dalam hal ini, nilai dari statistiknya akan banyak yang bernilai NaN karena perhitungannya menjadi inkonsisten.
# Ini diakibatkan benturan tipe data yang tidak saling berhubungan
```

```python
# Untuk mengambil data satuan yang lebih spesifik
# Adapun pendekatan yang saya ambil adalah menggunakan JSON (karena terbiasa)
# Untuk itu saya butuh library json (biasanya sudah build-in di environment python versi berapapun)
# Saya import "json" sebagai "json_lib" karena variabel "json" sudah saya gunakan untuk menampung data sebelumnya
import json as json_lib

# Saya load kembali response dari API Kaltim yang sudah saya knoversi kebentung text atau string
element = json_lib.loads(result.text)

# Kemudian, saya tentukan spesifik index dari dataset yang ada
print(element[2])
```

```python
# Terakhir untuk melakukan pencarian berdasarkan judul berita, saya menggunakan fungsi berikut
display(list(filter(lambda x:x["judul"]=="Pemkot Samarinda Kejar 300 Point untuk Raih Kategori Nindya Kota Layak Anak Tahun 2023", element))[0])
```
