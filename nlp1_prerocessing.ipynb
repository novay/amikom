#### **Step 3: Preprocessing**

```python
# %pip install nltk numpy sastrawi num2words

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from num2words import num2words
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

import numpy as np
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Mengubah semua text menjadi huruf kecil
def huruf_kecil(data):
    return np.char.lower(data)

def hapus_stopwords(data, lang):
    if(lang == 'id'): # Untuk Bahasa Indonesia gunakan Sastrawi
        stop_words = StopWordRemoverFactory().get_stop_words()    
    else: # Bila bukan, paksakan menggunakan stemming inggris dari NTLK
        stop_words = stopwords.words('english')
    
    words = word_tokenize(str(data))
    new_text = ""
    for w in words:
        if w not in stop_words and len(w) > 1:
            new_text = new_text + " " + w
    return new_text

def hapus_tanda_baca(data):
    symbols = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n"
    for i in range(len(symbols)):
        data = np.char.replace(data, symbols[i], ' ')
        data = np.char.replace(data, "  ", " ")
    data = np.char.replace(data, ',', '')
    return data

def hapus_penyingkat(data):
    return np.char.replace(data, "'", "")

def stemming(data, lang):
    if lang == 'id': # Untuk Indonesia gunakan Sastrawi
        stemmer = StemmerFactory().create_stemmer()
    else: # Untuk Inggris gunakan Bahasa Asing
        stemmer = PorterStemmer()
    
    tokens = word_tokenize(str(data))
    new_text = ""
    for w in tokens:
        new_text = new_text + " " + stemmer.stem(w)
    return new_text

# Mengubah angka menjadi string
def konversi_angka_ke_kata(data, lang):
    tokens = word_tokenize(str(data))
    new_text = ""
    for w in tokens:
        try:
            w = num2words(int(w), lang = lang)
        except:
            a = 0
        new_text = new_text + " " + w
    new_text = np.char.replace(new_text, "-", " ")
    
    return new_text

def preprocess(data, lang):
    data = huruf_kecil(data)
    data = hapus_tanda_baca(data)
    data = hapus_penyingkat(data)
    data = hapus_stopwords(data, lang)
    data = konversi_angka_ke_kata(data, lang)
    data = stemming(data, lang)
    
    # Khusus untuk Bahasa Inggris: karena perubahan angka ke kata
    if lang == 'en':
        data = hapus_tanda_baca(data) # Menghilangkan koma - "one hundred thousand, five hundred"
        data = hapus_stopwords(data, lang) # Menghapus stopwords "and" - "two thousand and twenty three"
    
    return data
```

```python
# Kondisi Penggunaan Data (Nonaktifkan salah satu untuk melanjutkan)

data_usage = new_result_indo
lang = 'id'

# data_usage = result_english
# lang = 'en'

######################################################################

# Extracting Data
N = len(data_usage)

# Fungsi untuk menghilangkan tags HTML
CLEANR = re.compile('<.*?>')
def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext

# Siapkan variabel untuk menampung token pasca preprocessing
processed_title = []
processed_text = []
for i in data_usage[:N]:
    processed_title.append(word_tokenize(str(preprocess(i['judul'], lang))))
    processed_text.append(word_tokenize(str(preprocess(cleanhtml(i['isi']), lang))))

processed_title
processed_text
```

#### **Step 4: Calculating TF-IDF**

##### Calculating DF for all words

```python
DF = {}
for i in range(N):
    tokens = processed_text[i]
    for w in tokens:
        try:
            DF[w].add(i)
        except:
            DF[w] = {i}

    tokens = processed_title[i]
    for w in tokens:
        try:
            DF[w].add(i)
        except:
            DF[w] = {i}

for i in DF:
    DF[i] = len(DF[i])
    
total_vocab_size = len(DF)
print(total_vocab_size)

total_vocab = [x for x in DF]
print(total_vocab[:100])
```

##### Calculating TF-IDF for body
We will consider this as the actual tf-idf as we will add the title weight to this.

```python
from collections import Counter

def doc_freq(word):
    c = 0
    try:
        c = DF[word]
    except:
        pass
    return c

doc = 0
tf_idf = {}
for i in range(N):
    tokens = processed_text[i]
    
    counter = Counter(tokens + processed_title[i])
    words_count = len(tokens + processed_title[i])
    
    for token in np.unique(tokens):
        tf = counter[token]/words_count
        df = doc_freq(token)
        idf = np.log((N+1)/(df+1))
        
        tf_idf[doc, token] = tf*idf
    doc += 1

tf_idf
```

##### Calculating TF-IDF for Title

```python
doc = 0
tf_idf_title = {}
for i in range(N):
    tokens = processed_title[i]
    counter = Counter(tokens + processed_text[i])
    words_count = len(tokens + processed_text[i])

    for token in np.unique(tokens):
        tf = counter[token]/words_count
        df = doc_freq(token)
        idf = np.log((N+1)/(df+1))
        
        tf_idf_title[doc, token] = tf*idf
    doc += 1

tf_idf_title
```
